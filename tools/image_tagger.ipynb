```
uploaded = google.colab.files.upload()

# Assumes only one file uploaded
for fn in uploaded.keys():
  a_filename = fn
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```
a_font = PIL.ImageFont.truetype('ChunkFive-Regular.otf', 50)
a_bigger_font = PIL.ImageFont.truetype('ChunkFive-Regular.otf', 65)

#blank_image = Image.new('RGBA', (400, 300), 'white')
in_image = PIL.Image.open(a_filename)
img_draw = PIL.ImageDraw.Draw(in_image)
#img_draw.rectangle((70, 50, 270, 200), outline='red', fill='blue')
#a_font = ont = ImageFont.truetype(font_path, font_size)

a_blackish = '#202020'
a_red = '#cc0808' # '#cc0009' # '#c81020'
a_green = '#00763a'
a_blue = '#0131A8'

img_draw.text((652, 75), 'Whiteboarder', fill=a_blackish, font=a_bigger_font)

img_draw.text((655, 188), ' •', fill=a_red, font=a_bigger_font)
img_draw.text((707, 196), 'A Two Bit Image Processor', fill=a_red, font=a_font)

img_draw.text((652, 301), ' •', fill=a_green, font=a_bigger_font)
img_draw.text((705, 309), 'Runs for Free on Colab', fill=a_green, font=a_font)

img_draw.text((654, 417), ' •', fill=a_blue, font=a_bigger_font)
img_draw.text((706, 426), 'MIT Licensed Python', fill=a_blue, font=a_font)
in_image.save('drawn_image.png')

pil_img = Image(filename='drawn_image.png')
display(pil_img)
```
